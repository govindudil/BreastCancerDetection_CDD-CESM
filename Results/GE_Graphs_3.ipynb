```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
def readCSV(file, columns_to_read):
  rows = []
  firstLine = True
  with open(file, 'r') as file:
      for line in file:
          if firstLine:
            firstLine = False
            continue
          row = []
          in_quotes = False
          field = ''
          for ch in line.strip():
              if ch == '"':
                  in_quotes = not in_quotes
                  continue
              if ch == ',' and not in_quotes:
                  row.append(field)
                  field = ''
              else:
                  field += ch
          row.append(field)
          rows.append(row)

  df = pd.DataFrame(rows, columns=columns_to_read)
  return df
```

```python
parent_folder_path = "/content/gdrive/MyDrive/sync/results"

# Get a list of folder names in the parent folder
folder_names = os.listdir(parent_folder_path)

AUC_train_best_array = []
AUC_val_best_array = []
AUC_train_avg_array = []
AUC_val_avg_array = []

for folder_name in folder_names:
    folder_path = f'{parent_folder_path}/{folder_name}'

    # Get a list of file names in the folder
    file_names = os.listdir(folder_path)
    # print(file_names)

    train_df = []
    val_df = []
    best_df = []
    AUC_df_train = pd.DataFrame(columns= range(25))
    AUC_df_val = pd.DataFrame(columns= range(25))
    AUC_df_train_avg = pd.DataFrame(columns= range(25))
    AUC_df_val_avg = pd.DataFrame(columns= range(25))

    columns_to_read = ["Generation",	"FPR",	"1 - TPR",	"1 - AUC",	"F1",	"BA",	"MCC",	"Accuracy",	"TP",	"FP",	"TN",	"FN", "Phenotype"]

    count = 0
    for file_name in file_names:
        print(file_name)
        data_df_1 = []
        data_df_2 = []
        best_individuals_train = []
        best_individuals_val = []
        columns = ["Generation",	"FPR",	"1 - TPR",	"1 - AUC",	"F1",	"BA",	"MCC",	"Accuracy",	"TP",	"FP",	"TN",	"FN", "Phenotype", "Distance", "Fold"]
        best_individuals_val = pd.DataFrame(columns=columns)
        best_individuals_train = pd.DataFrame(columns=columns)
        for fold in range(5):
            df1 = readCSV(f'{folder_path}/{file_name}/best_train_indvidual_{fold+1}.csv', columns_to_read)
            df1["FPR"] = df1["FPR"].astype(float)
            df1["1 - TPR"] = df1["1 - TPR"].astype(float)
            df1["1 - AUC"] = df1["1 - AUC"].astype(float)
            df1["Distance"] = df1["FPR"]**2 + df1["1 - TPR"]**2 + df1["1 - AUC"]**2
            df1["Fold"] = fold

            df2 = readCSV(f'{folder_path}/{file_name}/best_val_indvidual_{fold+1}.csv', columns_to_read)
            df2["FPR"] = df2["FPR"].astype(float)
            df2["1 - TPR"] = df2["1 - TPR"].astype(float)
            df2["1 - AUC"] = df2["1 - AUC"].astype(float)
            df2["Distance"] = df2["FPR"]**2 + df2["1 - TPR"]**2 + df2["1 - AUC"]**2
            df2["Fold"] = fold

            data_df_1.append(df1)
            data_df_2.append(df2)
        for generation in range(len(data_df_1[0])):
            best_individuals_val.loc[generation] = data_df_2[0].loc[generation]
            best_individuals_train.loc[generation] = data_df_1[0].loc[generation]
            for fold in range(5):
                if best_individuals_val.loc[generation, "Distance"] > data_df_2[fold].loc[generation, "Distance"]:
                    best_individuals_val.loc[generation] = data_df_2[fold].loc[generation]
                if best_individuals_train.loc[generation, "Distance"] > data_df_1[fold].loc[generation, "Distance"]:
                    best_individuals_train.loc[generation] = data_df_1[fold].loc[generation]
        train_df.append(data_df_1)
        val_df.append(data_df_2)
        best_df.append(best_individuals_val)
        AUC_df_val[count] = 1 - best_individuals_val['1 - AUC']
        AUC_df_train[count] = 1 - best_individuals_train['1 - AUC']
        AUC_df_val_avg[count] = 1 - (data_df_2[0].loc[:, "1 - AUC"] + data_df_2[1].loc[:, "1 - AUC"] + data_df_2[2].loc[:, "1 - AUC"] + data_df_2[3].loc[:, "1 - AUC"] + data_df_2[4].loc[:, "1 - AUC"])/5
        AUC_df_train_avg[count] = 1 - (data_df_1[0].loc[:, "1 - AUC"] + data_df_1[1].loc[:, "1 - AUC"] + data_df_1[2].loc[:, "1 - AUC"] + data_df_1[3].loc[:, "1 - AUC"] + data_df_1[4].loc[:, "1 - AUC"])/5
        count = count + 1
    AUC_val_best_array.append(AUC_df_val)
    AUC_train_best_array.append(AUC_df_train)
    AUC_val_avg_array.append(AUC_df_val_avg)
    AUC_train_avg_array.append(AUC_df_train_avg)

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
iterations = np.arange(len(AUC_df))
min_values = AUC_df.min(axis = 1)
max_values = AUC_df.max(axis = 1)
mean_values = AUC_df.mean(axis = 1)
error_values = [mean_values - min_values, max_values - mean_values] # [lower error, upper error]

# Plotting
plt.figure(figsize=(10, 6))

plt.errorbar(iterations, mean_values, yerr=error_values, fmt='o', ecolor='gray', elinewidth=1, capsize=2)

plt.xlabel('Generations')
plt.ylabel('AUC')
plt.title('Best validation AUC')
plt.show()
```
